```python
# ==========================================
# TESLA AUTOPILOT CLONE
# OBJECT DETECTION USING YOLOv8
# ==========================================

# Install Dependencies
!pip install -q ultralytics opencv-python

# Import Libraries
import cv2
import numpy as np
from ultralytics import YOLO
from google.colab.patches import cv2_imshow

print("="*60)
print("TESLA AUTOPILOT CLONE - OBJECT DETECTION")
print("="*60)

# ==========================================
# LOAD YOLOv8 MODEL
# ==========================================

model = YOLO("yolov8n.pt")

print("YOLOv8 Model Loaded Successfully")

# ==========================================
# INPUT VIDEO
# ==========================================

video_path = "road_video.mp4"

cap = cv2.VideoCapture(video_path)

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# ==========================================
# OUTPUT VIDEO
# ==========================================

output_path = "detected_output.mp4"

fourcc = cv2.VideoWriter_fourcc(*'mp4v')

out = cv2.VideoWriter(
    output_path,
    fourcc,
    fps,
    (width, height)
)

frame_count = 0

# ==========================================
# PROCESS VIDEO
# ==========================================

while True:

    success, frame = cap.read()

    if not success:
        break

    frame_count += 1

    # Run Detection
    results = model(frame)

    annotated_frame = results[0].plot()

    out.write(annotated_frame)

    if frame_count % 100 == 0:
        print(
            f"Processed {frame_count} frames..."
        )

cap.release()
out.release()

print("\nVideo Processing Completed")
print("Output Saved:", output_path)

# ==========================================
# DISPLAY SAMPLE FRAME
# ==========================================

cap = cv2.VideoCapture(output_path)

ret, frame = cap.read()

if ret:

    print("\nSample Detection Output")

    cv2_imshow(frame)

cap.release()

# ==========================================
# IMAGE TESTING
# ==========================================

image_path = "test_image.jpg"

try:

    image = cv2.imread(image_path)

    results = model(image)

    detected_image = results[0].plot()

    print("\nImage Detection Result")

    cv2_imshow(detected_image)

except:
    print("No test image found")

# ==========================================
# OBJECT COUNTING
# ==========================================

image = cv2.imread(image_path)

if image is not None:

    results = model(image)

    boxes = results[0].boxes

    print("\nDetected Objects")

    object_count = {}

    for box in boxes:

        cls = int(box.cls[0])

        name = model.names[cls]

        object_count[name] = (
            object_count.get(name, 0) + 1
        )

    for obj, count in object_count.items():

        print(
            f"{obj}: {count}"
        )

# ==========================================
# PERFORMANCE SUMMARY
# ==========================================

print("\nProject Completed Successfully")
```
